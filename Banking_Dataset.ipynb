**E. Calculated Columns / Transformations (10 Questions)**

1.	Add Total_Balance = Balance + Fixed_Deposit + Recurring_Deposit.

```python
df['Total_Balance']=df['Balance'] + df['Fixed_Deposit'] + df['Recurring_Deposit']
```

2.	Add Loan_Status flag (1 if Loan_Amount > 0 else 0).

```python
df['Loan_Status']=df['Loan_Amount']
for i in df.index:
  if df.loc[i,'Loan_Status'] > 0:
    df.loc[i,'Loan_Staus'] = '1'
  else:
        df.loc[i,'Loan_Staus'] = '0'

```

3.	Add Customer_Tenure = Current_Date − Account_Open_Date.

```python
from datetime import date
df['Account_Open_Date'] = pd.to_datetime(df['Account_Open_Date'], format='%d-%m-%y', errors='coerce')
df['Customer_Tenure'] = (pd.Timestamp.now() - df['Account_Open_Date']).dt.days
```

4.	Create Account_Category (Low Balance, Medium, High Balance).

```python
df['Account_Category'] = df['Balance']
x=df['Balance'].median()
for i in df.index:
  if df.loc[i,'Account_Category'] < 2000000:
    df.loc[i,'Account_Category'] = 'Low Balance'
  elif df.loc[i,'Account_Category'] <= x:
    df.loc[i,'Account_Category'] = 'Median Balance'
  elif df.loc[i,'Account_Category'] > x:
    df.loc[i,'Account_Category'] = 'High Balance'
```
